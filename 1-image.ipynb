# Images as data

```python
import matplotlib.pyplot as plt
```

```python
data = plt.imread('lola.jpeg')
```

```python
plt.imshow(data)
plt.show()
```

# Modifying image data

```python
data[:, :, 1] = 0
data[:, :, 2] = 0
plt.imshow(data)
plt.show()
```

# Changing image data

```python
data[500:750, 450:650, :] = [0,1,0]
plt.imshow(data)
plt.show()
```
